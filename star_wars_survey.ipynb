## Reading in the data

```python
import pandas as pd

star_wars = pd.read_csv('StarWars.csv', encoding='ISO-8859-1')
star_wars.head(10)
star_wars.columns
```

```python
# remove any rows where 'RespondentID' is NaN

star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

## Converting 'yes/no' columns to Boolean values

```python
# mapping dictionary

yes_no = {
    'Yes': True,
    'No': False
}

# columns to convert

cols = [
    'Have you seen any of the 6 films in the Star Wars franchise?',
    'Do you consider yourself to be a fan of the Star Wars film franchise?'
]

for c in cols:
    star_wars[c] = star_wars[c].map(yes_no)
    
star_wars.head()
```

```python
import numpy as np

# mapping to convert responses relating to which movies respondents have seen to Boolean values

seen = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
}

for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(seen)
    
# give 'seen' columns more intuitive names

num = 1
for c in star_wars.columns[3:9]:
    star_wars.rename(columns={c : 'seen_{}'.format(num)}, inplace=True)
    num += 1

star_wars.head()
```

## Convert rankings columns to numeric format

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# give rankings columns more descriptive names

num = 1
for c in star_wars.columns[9:15]:
    star_wars.rename(columns={c : 'ranking_{}'.format(num)}, inplace=True)
    num += 1

star_wars.iloc[:,9:15].head()
```

## Calculate mean rankings

```python
mean_ranking = {}

cols = [
    'ranking_1',
    'ranking_2',
    'ranking_3',
    'ranking_4',
    'ranking_5',
    'ranking_6'
]

for c in cols:
    mean = star_wars[c].mean()
    mean_ranking[c] = mean
    
display(mean_ranking)
```

```python
import matplotlib.pyplot as plt

index = np.arange(6)
width = 0.5
x_ticklabels = ['Ep. I','Ep. II','Ep. III','Ep. IV','Ep. V','Ep. VI']

fig, ax = plt.subplots()
bars = ax.bar(index, mean_ranking.values(), width)

ax.set_ylabel('Ranking')
ax.set_title('Mean ranking of each film')
ax.set_xticks(index)
ax.set_xticklabels(x_ticklabels)

plt.show()
```

On average, the original trilogy is preferred to the newer one, with 'Episode V: The Empire Strikes Back' the most highly ranked.

## Investigate viewing counts

```python
seen_count = star_wars[star_wars.columns[3:9]].sum()

index = np.arange(6)
width = 0.5

fig, ax = plt.subplots()
bars = ax.bar(index, seen_count, width)

ax.set_ylabel('Count')
ax.set_title('Number of respondents that have seen film')
ax.set_xticks(index)
ax.set_xticklabels(x_ticklabels)

plt.show()
```

Episodes IV and V were the most and second most viewed films respectively, which partly explains why they were the most and second most highly ranked respectively. Reinforces the idea that the original trilogy is more popular.

## Comparing ranking and viewership statistics between male and female respondents

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']

males_mean_ranking = males[males.columns[9:15]].mean()
females_mean_ranking = females[females.columns[9:15]].mean()

males_seen_count = males[males.columns[3:9]].sum()
females_seen_count = females[females.columns[3:9]].sum()
```

```python
index = np.arange(6)
width = 0.25

fig, ax = plt.subplots()
bars1 = ax.bar(index, males_mean_ranking, width, color='r')
bars2 = ax.bar(index + width, females_mean_ranking, width, color='y')

ax.set_ylabel('Ranking')
ax.set_title('Mean ranking of each film')
ax.set_xticks(index + width / 2)
ax.set_xticklabels(x_ticklabels)

ax.legend((bars1[0], bars2[0]), ('Males', 'Females'))

plt.show()
```

```python
index = np.arange(6)
width = 0.25

fig, ax = plt.subplots()
bars1 = ax.bar(index, males_seen_count, width, color='r')
bars2 = ax.bar(index + width, females_seen_count, width, color='y')

ax.set_ylabel('Count')
ax.set_title('Number of respondents that have seen film')
ax.set_xticks(index + width / 2)
ax.set_xticklabels(x_ticklabels)

ax.legend((bars1[0], bars2[0]), ('Males', 'Females'))

plt.show()
```
